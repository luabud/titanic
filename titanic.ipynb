```python
## read the titanic-dataset
import pandas as pd

x = pd.read_csv('Titanic-Dataset.csv')
x.head()
```

```python
## visualize the number of survivors by age
import matplotlib.pyplot as plt

x['Age'].hist(bins=16)
plt.show()
```

```python
# plot the correlation between fare and survival
import seaborn as sns
sns.barplot(x='Survived', y='Fare', data=x)
plt.show()

```

```python
# plot the correlation between sex and survival
import seaborn as sns
sns.barplot(x='Survived', y='Sex', data=x)
plt.show()
```

```python
# plot the correlation between survived and class
import seaborn as sns
sns.barplot(x='Survived', y='Pclass', data=x)
plt.show()
```
